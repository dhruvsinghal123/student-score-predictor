```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

url= r"C:\Users\hp\Downloads\student_scores.csv"
data= pd.read_csv(url)
print(data.head())
plt.scatter(data["Hours"],data["Scores"],color="blue")
plt.title("Exam Scores Prediction")
plt.xlabel("Hours Studied")
plt.ylabel("Scored")
plt.grid(True)
plt.show()
# it will split the data into test and train 
x= data[['Hours']] #input 
y= data[['Scores']]  # output 
x_train , x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state= 42)
# now we will train the data using the linear regression 
model= LinearRegression()
model.fit(x_train , y_train)
# Now test the model 
# Predict on test set
y_pred = model.predict(x_test)
# Ensure both are 1D arrays
y_test_1d = y_test.values.ravel()  # In case y_test is a DataFrame
y_pred_1d = np.ravel(y_pred)       # In case y_pred is 2D
# Create DataFrame
df = pd.DataFrame({
    'Actual': y_test_1d,
    'Predicted': y_pred_1d
})

print(df)
mse= mean_squared_error(y_test,y_pred)
print(f"Mean Squared error: {mse:2f}")

# Predict for a custom input
hours = 8.5
predicted_score = model.predict([[hours]])
print(f"If a student studies {hours} hours, they are expected to score {float(predicted_score[0]):.2f}")
data
```
